# GMM (Gaussian Mixture Model)


```python
from sklearn.datasets import make_classification   # 분류 문제에 최적화된 데이터를 만들어주는 라이브러리 
import matplotlib.pyplot as plt 


X, _ = make_classification(
    n_samples=300,              # 데이터 개수 
    n_features=2,               # 특성 개수 
    n_informative=2,            # 유의미한 특성 개수 
    n_redundant=0,              # 중복 특성 개수
    n_clusters_per_class=1,     # 클래스 당 클러스터 수 
    n_classes=2,                # 클래스(레이블) 개수
    random_state=42
)

plt.scatter(X[:, 0], X[:, 1], s=50, c='gray', marker='o', edgecolors='k')
plt.title('scikit-learn making data')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=2, random_state=42)           # 군집별 분류 시각화
gmm.fit(X)

labels = gmm.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, edgecolors='k')
plt.title('GMM cluster result')
plt.show()
```

```python
probs = gmm.predict_proba(X)                                       # 확률값에 의해 표현됨(어떠한 군집에 포함될 확률이 높다, 낮다로 표현. 헷갈리지 말기!!) - 군집의 경계에 있는 값의 경우, 확률이 낮은 것을 몰수 있음. 

plt.scatter(X[:, 0], X[:, 1], c=probs.max(axis=1), s=50, edgecolors='k')   # c=probs.max 컬러 조건. 가우시안은 확률을 반환하기 때문에 값이 두개(각 군집당 하나씩) 그중 큰값으로 군집화 
plt.title('GMM cluster rate')
plt.colorbar(label='rate')
plt.show()
```
