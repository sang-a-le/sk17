# series 생성

```python
# python list 활용 -> Series 생성 (1차원의 배열 형태)
teachers = ['squirrel', 'rabbit', 'tiger', 'monkey', 'treefrog']
teachers_ser = pd.Series(teachers)   # series 앞 s 대문자 (앞글자가 대문자이면 class)

print(type(teachers_ser))   # <class 'pandas.core.series.Series'> 사용자 정의 series 자료형
print(teachers_ser[0])

print(teachers_ser)   # 1차원임. 인덱스랑 착각해서 2차원으로 생각하면 안됨. 
                      # pandas에서 나오는 type 중 문자열은 object로 표현

teachers_ser.name = '오지라퍼스'        # 인스턴스 속성 정하기
print(teachers_ser)
```

```python
# Numpy ndarray 활용 -> Series 생성
nums_ser = pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])   #index : 키워드 인자 (인덱스 지정) - 개수 맞춰주기 & 유니크한 값

print(nums_ser)
print(nums_ser[1])  # 임의 부여된 인덱스. 값이 프린트가 되긴 함. (future version)
print(nums_ser['b'])
print(nums_ser.iloc[2], nums_ser.loc['c'])              # 매서드는 .매서드명() / 속성 부여는 .속성명[]
         # iloc (integer location) : 인덱스(번호)를 통한 참조
         # loc (location) : 인덱스 라벨을 통한 참조(지정해준 문자-이름)
```

```python
# dictionary 활용 -> Series 생성
info = {
    'a' : 10, 
    'b' : 20, 
    'c' : 30
}                 # 딕셔너리 생성

info_ser = pd.Series(info)   # 키 값이 라벨, value가 값
print(info_ser)

info_ser.index = ['A', 'B', 'C']
print(info_ser)

# info2 = {'a':(10, 20, 30), 'b':(40, 50, 60)}
# info_ser2 = pd.Series(info2)
# print(info_ser2)
```

```python
# scalar value 활용 -> Series 생성
num_ser = pd.Series(7.16)
num_ser = pd.Series(7.16, index=['ㄱ', 'ㄴ', 'ㄷ', 'ㄹ', 'ㅁ'])   # 스칼라 값은 단일값이기 때문에 인덱스 개수만큼 늘어나게 만들 수 있음. (array는 수가 동일해야함.)
print(num_ser)
```

### Series 속성

```python
movies_ser = pd.Series(['F1 더 무비', '쥬라기 월드', '슈퍼맨', '노이즈', '괴기열차'])
movies_ser
```

```python
movies_ser.values        #값만 반환 (array 형태)

print(movies_ser.values)
print(type(movies_ser.values))
```

```python
print(movies_ser.array)
print(type(movies_ser.array))
```

```python
movies_ser.index         # index를 별도 지정하지 않은 경우 기본적으로 숫자 인덱스

movies_ser.index = ['1st', '2nd', '3rd', '4th', '5th']
movies_ser.index         # index를 지정한 경우 라벨 인덱스 
```

```python
# series의 값이 모두 고유한 값인지 여부 (T: 중복값이 없음 / F: 중복값이 있음) : 모든 값에 대한 반환
movies_ser.is_unique   # is 가 붙으면 bool 반환. 
```

```python
print(movies_ser.dtype) # 요소의 자료형
print(movies_ser.shape) # 시리즈의 형태 
print(movies_ser.ndim) # 시리즈 차원의 깊이
print(movies_ser.size) # 시리즈 요소의 개수 
```

### Series 메서드

```python
nums_ser = pd.Series([2025, 7, 16, 11, 2])
```

# S&P500 데이터 활용

```python
sp_500_df = pd.read_csv('./data/s_p500_Prices.csv')
sp_500_df
```

```python
type(sp_500_df)
```

```python
# df.squeeze() : DataFrame의 Series가 하나인 경우 Series 객체 반환
sp_500_ser = sp_500_df.squeeze()  # np.squeeze = 차원제거 / 여기 스퀴즈는 데이터프레임에서 제공하는 메서드를 뜻함.
print(sp_500_df)
```

```python
sp_500_ser.info()
```

```python
sp_500_ser.describe()
```

```python
print(sp_500_ser.count())
print(sp_500_ser.min())
print(sp_500_ser.median())
print(sp_500_ser.max())
print(sp_500_ser.mean())
print(sp_500_ser.std())
print(sp_500_ser.var())
```

```python
# 인덱스 번호로 인덱싱 & 슬라이싱

# 인덱스 50의 값 출력
# 인덱스 100~200 사이의 값 출력
print(sp_500_ser[50])
print(sp_500_ser[100:201])
```

```python
# fancy indexing

# 1000번 index, 2000번 index의 값 동시 출력
print(sp_500_ser[[1000, 2000]])
```

```python
# boolean indexing

# 값이 3000보다 큰 값 출력
print(sp_500_ser[sp_500_ser > 3000])

print(sp_500_ser[sp_500_ser.values > 3000]) # 값 중에서 반환

for temp in sp_500_ser[sp_500_ser > 3000] :
    print(sp_500_ser)
```

```python
# in 연산: 특정 (무엇)을 가지고 있는지 여부 
print(1416.51001 in sp_500_ser.values)    # in 연산. 특정 값을 가지고 있는지 여부 

# .values를 붙이지 않으면 기본적으로 index로 찾음
print(1416.51001 in sp_500_ser)           # false
print(2000 in sp_500_ser)                 # true
```

```python
# 정렬
print(sp_500_ser.sort_values())   # sort_index : 인덱스에 대한 정렬.(이미 정렬되어 있는 상태이기 때문에 라벨 인덱스를 부여해준 경우가 아니라면 무의미)
print(sp_500_ser)     # sort 는 정렬된 결과를 반환만 하고, 저장하지 않음. 
```

```python
sp_500_ser = sp_500_ser.sort_values(ascending=False)  #내림차순 정렬 지원
print(sp_500_ser)
```
