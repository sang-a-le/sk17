### 텍스트 전처리

```python
import pandas as pd 
import numpy as np 
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
```

```python
spam_df = pd.read_csv('./sms_spam.csv')
spam_df.head()
```

```python
corpus = spam_df['CN'][0]  # 스팸데이터 중 0번째만 반환
corpus
# 전체 데이터에 대해 워드클라우드를 하고 싶으면 전체 글자를 corpus로 반환
```

```python
okt = Okt()
nouns = okt.nouns(corpus) # 명사만 추출 
nouns
```

```python
word_count = Counter(nouns) # 빈도 계산
word_count
```

```python
# 불용어 처리 
ko_stopword = ['및', '더', '수']
word_count = {word: count for word, count in word_count.items() if word not in ko_stopword}
word_count
```

```python
# 워드 클라우드 만들기 
word_cloud = WordCloud(
    width=800, 
    height=800, 
    background_color='white',
    font_path='C:\\Windows\\Fonts\\batang.ttc'
).generate_from_frequencies(word_count)  # word_count 자료의 빈도에 기반해서 위의 설정으로 만들어라 

word_cloud
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```

```python
# 전처리 전 데이터로 워드클라우드 
text = spam_df['CN'][0]

word_count2 = Counter(text)

# 워드 클라우드 만들기 
word_cloud = WordCloud(
    width=800, 
    height=800, 
    background_color='white',   # 기본값은 black
    font_path='C:\\Windows\\Fonts\\batang.ttc'
).generate_from_frequencies(word_count2)  # word_count 자료의 빈도에 기반해서 위의 설정으로 만들어라 

plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```

### 전체 데이터를 corpus로 Word cloud 그리기

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# corpus = spam_df['CN']
# text = " ".join(map(str, corpus))
# corpus_token = word_tokenize(text)
# corpus_token

okt = Okt()
# nouns = okt.nouns(corpus_token) # 명사만 추출 
# nouns

corpus = spam_df['CN']

nouns = []

for temp in corpus:
    nouns.extend(okt.nouns(temp))

nouns
```

```python
word_count = Counter(nouns) # 빈도 계산
```

```python
# 불용어 처리 
ko_stopword = ['및', '더', '수']
word_count = {word: count for word, count in word_count.items() if word not in ko_stopword}
word_count
```

```python
from PIL import Image # 파이썬 레벨의 이미지 로드 지원

# 워드 클라우드 만들기 
word_cloud = WordCloud(
    width=800, 
    height=800, 
    background_color='white',
    font_path='C:\\Windows\\Fonts\\batang.ttc',
    mask = np.array(Image.open('C:\\skn_17\\nlp\\02_preprocessing\\image.jpg'))  # 이미지 모양 안에 맞게 만들어짐
).generate_from_frequencies(word_count)  # word_count 자료의 빈도에 기반해서 위의 설정으로 만들어라 

word_cloud
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```
